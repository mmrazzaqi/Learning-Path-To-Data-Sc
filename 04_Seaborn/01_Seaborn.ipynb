```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# matplolib default parameter

plt.figure()
x1=[9, 15, 6, 10, 4]
x2=[4, 9, 8, 14, 6]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
# seaborn default parameter

import seaborn as sns
```

```python
sns.set()
plt.figure()
x1=[9, 15, 6, 10, 4]
x2=[4, 9, 8, 14, 6]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
# to contro style, seaborn provides two methods
# set_style(style)   (set aesthetic style of the plot)
# axes_style()

sns.set_style('whitegrid')     # 'whitegrid' is pre configured style (it is dict of parameters)
plt.figure()
x1=[9, 15, 6, 10, 4]
x2=[4, 9, 8, 14, 6]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
# parameter mapings to override the values in preset style dictionary

sns.set()
plt.figure()
x1=[9, 15, 6, 10, 4]
x2=[4, 9, 8, 14, 6]
with sns.axes_style("dark"):
    plt.plot(x1, label="Group A")
    plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
# to control the context, we use two functions
# sns.set_context(context, [font_scale], [rc])

# it scale the elements of figure (zoom the graph)

sns.set_context("poster")
plt.figure()
x1=[9, 15, 6, 10, 4]
x2=[4, 9, 8, 14, 6]
plt.plot(x1, label="Group A")
plt.plot(x2, label="Group B")
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("scores.csv")
```

```python
data.head()
```

```python
grpA = data[data.columns[0]].tolist()
grpb = data[data.columns[1]].tolist()
```

```python
df = pd.DataFrame({'Groups':['Group A']*len(grpA) + ['Group B']*len(grpb),
                  'IQ Score': grpA + grpb})
```

```python
plt.figure(dpi=100)
sns.set_style('whitegrid')
sns.boxplot(x='Groups', y='IQ Score', data=df)
sns.despine(left=True, right=True, top=True)
plt.show()
```

### Color Palettes

```python
# deep color palette

palette_deep = sns.color_palette('deep')
sns.palplot(palette_deep)
```

```python
# muted color palette

palette_muted = sns.color_palette('muted')
sns.palplot(palette_muted)
```

```python
# dark color palette

palette_dark = sns.color_palette('dark')
sns.palplot(palette_dark)
```

```python
# bright color palette

palette_bright = sns.color_palette('bright')
sns.palplot(palette_bright)
```

```python
# pastel color palette

palette_pastel = sns.color_palette('pastel')
sns.palplot(palette_pastel)
```

```python
# colorblind color palette

palette_colorblind = sns.color_palette('colorblind')
sns.palplot(palette_colorblind)
```

### Sequential & Diverge color palette

```python
# sequential palette use at "relatively low or uninteresting" and relatively high or interesting

palette_sq_1 = sns.light_palette('brown')
sns.palplot(palette_sq_1)
```

```python
# reverse version

palette_sq_2 = sns.light_palette('brown', reverse=True)
sns.palplot(palette_sq_2)
```

```python
# div palette
# show relative increase or decrease

palette_div = sns.color_palette('coolwarm', n_colors=7)
sns.palplot(palette_div)
```

```python
# custom div CP

pal_div_custom = sns.diverging_palette(440, 40, n=7)
sns.palplot(pal_div_custom)
```
