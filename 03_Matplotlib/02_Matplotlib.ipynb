### Figure Customization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1,9)
```

```python
y = np.arange(3,27,3)
```

```python
# add fig and 4 axes to it

fig, axes = plt.subplots(nrows=2, ncols=1)

# to make more readble

plt.tight_layout()
```

```python
print(type(axes))
```

```python
# we can acees axes array elements(axes)

print(axes[1])
```

```python
# we can iterate through axes array

for ax in axes:
    print(ax.plot(x,y))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x,y)
axes[0].set_title("X-Y Graph")
axes[1].plot(x,x**2)
axes[1].set_title("X-X Square Graph")
plt.tight_layout()
```

```python
# define another fig

fig_2 = plt.figure(figsize=(6,4), facecolor="red")
axes_2=fig_2.add_axes([0,0,1,1])
axes_2.plot(x,x**3, color="orange")
```

```python
# define another fig

fig_3, axes_3 = plt.subplots(nrows=1, ncols=2, figsize=(14,6))
axes_3[0].plot(y,y**2, "black")
axes_3[0].set_title("Y Graph")
axes_3[1].plot(x,x**3, "green")
axes_3[1].set_title("X Graph")
```

```python
# save fig in PNG or PDF with savefig()

fig_3.savefig("fig_3.png")

# fig_3.savefig("fig_3.pdf")
```

```python
fig_4, axes_4 = plt.subplots(figsize=(14,6))
axes_4 = fig_4.add_axes([0,0,1,1])

axes_4.plot(x,x**0.2, label="X square root", color="red")
axes_4.plot(x,x**3, label="X Cube", color="black")
axes_4.plot(x,y, label="X-Y Graph", color="green")
axes_4.plot(y,y**2, label="X-Sinx", color="yellow")
axes_4.legend()
```

### Plot customization

```python
x_2 = np.random.randint(1,12,5)
y_2 = np.random.randint(1,10,5)
```

```python
fig_5, axes_5 = plt.subplots()
axes_5 = fig_5.add_axes([0,0,1,1])

# plot parameters (color, linewidth(lw), linestyle(ls), marker, markersize, markerfacecolor, 
#                  markeredgecolor, markeredgewidth)

axes_5.plot(x_2, y_2**2, color="red", lw=4, linestyle="--", marker="D", markersize=10, markerfacecolor="blue",
           markeredgecolor="yellow", markeredgewidth="2")


# axis limit (set_xlim   &  set_ylim  methd)

axes_5.set_xlim(0,20)
axes_5.set_ylim(0,100)
```

### Ticks (xticks, yticks) 
###  spines
### grid

```python
import math
```

```python
xx = np.arange(0, math.pi*2, 0.09)
yy = np.sin(xx)

fig_6 = plt.figure()
axes_6 = fig_6.add_axes([0,0,1,1])
axes_6.plot(xx,yy)

# we can add ticks
axes_6.set_xticks([0,2,4,6])
axes_6.set_yticks([-1,0,1])

# we ca aso add ticks label
axes_6.set_xticklabels(["zero", "two", "four", "six"])

# spines
# we can remove spines
axes_6.spines["right"].set_visible(False)
axes_6.spines["top"].set_visible(False)


# grid
axes_6.grid(color="red", ls="-.", lw=2)
```
