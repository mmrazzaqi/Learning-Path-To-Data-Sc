```python
# contacts data frame

contacts = [["mohsin", "mmrazzaqi@gmail.com", "1111111111"],
           ["razzaqi", "r@gmail.com", "2222222222"],
           ["ibrahim", "ab@gmail.com", "3333333333"]]

dataframe_6 = pd.DataFrame(contacts, columns = ["name", "email", "phone"])
dataframe_6
```

```python
# import regular expression liberary

import re
```

```python
# define maping function

def get_formatted_phone(value):
    result = re.fullmatch(r"(\d{3})(\d{3})(\d{4})", value)
    return "-".join(result.groups()) if result else value
```

```python
# call function

formatted_phones = dataframe_6["phone"].map(get_formatted_phone)
```

```python
formatted_phones
```

```python
dataframe_6["phone"] = formatted_phones
```

```python
dataframe_6
```

```python
join_right = pd.DataFrame({"Apple":['Appfel', 'Mela', 'Manzana', 'Pomne'],
                          "Orange":['Orange', 'Arancia', 'Naranja', 'Orange'],
                          "Banana":['Banana', 'Banana', 'Platano', 'Banana']},
                         index=['German', 'Italian', 'Spanish', 'French'])
join_right
```

```python
join_left = pd.DataFrame({"One":['Ein', 'Uno', 'Una', 'Unus','Un'],
                          "Two":['Zwei', 'Duo', 'Dos', 'Deux', 'Duo'],
                          "Three":['Two', 'Tre', 'Tres', 'Two', 'Tribus']},
                         index=['German', 'Italian', 'Spanish', 'French', 'Latin'])
join_left
```

```python
# join right

result_1 = join_right.join(join_left)
result_1
```

```python
# join left

result_2 = join_left.join(join_right)
result_2
```

```python
join_right = pd.DataFrame({"Apple":['Appfel', 'Mela', 'Manzana', 'Pomne'],
                          "Orange":['Orange', 'Arancia', 'Naranja', 'Orange'],
                          "Banana":['Banana', 'Banana', 'Platano', 'Banana'],
                         "key":['German', 'Italian', 'Spanish', 'French']})
join_right
```

```python
result_3 = join_right.join(join_left, on = "key")
result_3
```
