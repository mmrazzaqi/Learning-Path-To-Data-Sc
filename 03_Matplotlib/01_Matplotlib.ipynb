```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_values = np.arange(1,9)
x_values
```

```python
y_values = np.arange(1,24,3)
y_values
```

```python
plt.plot(x_values,y_values, color="orange")
plt.xlabel("x_values")
plt.ylabel("y_values")
plt.title("Plot")
plt.show()
```

```python
x_values_2 = np.linspace(-4,4,50)
x_values_2
```

```python
y_values_2 = x_values_2 ** 2
y_values_2
```

```python
plt.plot(x_values_2,y_values_2)
plt.show()
```

```python
# we can also use dot(.) or dash(-)

plt.plot(x_values_2,y_values_2, ".")
plt.show()
```

```python
# we can change color of dots

plt.plot(x_values_2,y_values_2, "r.")
plt.show()
```

```python
# multiple plots

plt.plot(x_values_2, np.sin(x_values_2), "r.")
plt.plot(x_values_2, np.cos(x_values_2), "g--")
plt.show()
```

```python
# bar plot

plt.bar(x_values, y_values)
plt.show()
```

```python
# bar plot horizental

plt.barh(x_values, y_values)
plt.show()
```

```python
# step plot

plt.step(x_values, y_values)
plt.show()
```

```python
# polar plot

plt.polar(x_values, y_values)
plt.show()
```

### Figure, Subplot and Axes

```python
x_values_4 = np.arange(-1,5,0.1)
x_values_4
```

```python
y_values_4 = np.exp(-x_values_4)*np.cos(5*x_values_4)
```

```python
plt.plot(x_values_4,y_values_4)
plt.show()
```

```python
# figure()

plt.figure(figsize= (10,8))
plt.plot(x_values_4,y_values_4)
```

```python
# with help of dpi, we can change dot per inch
# stackplot

plt.figure(dpi=(150))
plt.stackplot(x_values_4,y_values_4)
```

```python
# with facecolor, we can change figure rectangle color

plt.figure(facecolor="r")
plt.plot(x_values_4,y_values_4)
```
